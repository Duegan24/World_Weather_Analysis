```python
# Import the dependencies.
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import requests, time
from datetime import datetime
import sys
import gmaps
import gmaps.datasets
from config import g_api_key
```

```python
# Import WeatherPY_challenge.csv in dataframe
file_to_import = os.path.join("Analysis","WeatherPY_vacation.csv")

vacation_cities_df = pd.read_csv(file_to_import)
vacation_cities_df
```

```python
# Filter for each city:
city_1 = vacation_cities_df[vacation_cities_df["City"]=="Itarema"]
city_2 = vacation_cities_df[vacation_cities_df["City"]=="Humberto De Campos"]
city_3 = vacation_cities_df[vacation_cities_df["City"]=="Cururupu"]
city_4 = vacation_cities_df[vacation_cities_df["City"]=="Carutapera"]

#Filter for hotel specifics to display for markers
hotel_df = vacation_cities_df[vacation_cities_df["City"].isin(["Itarema","Humberto De Campos","Cururupu","Carutapera"])]
hotel_df
```

```python
# Configure gmaps to use your Google API key.
gmaps.configure(api_key=g_api_key)

# Latitude-longitude pairs
city_1_coord = (city_1.iloc[0]["Lat"],city_1.iloc[0]["Lng"])
city_2_coord = (city_2.iloc[0]["Lat"],city_2.iloc[0]["Lng"])
city_3_coord = (city_3.iloc[0]["Lat"],city_3.iloc[0]["Lng"])
city_4_coord = (city_4.iloc[0]["Lat"],city_4.iloc[0]["Lng"])


# Setting up the info_box_template
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Weather Description} and {Max Temp} °F</dd>
</dl>
"""

# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]

# Add locations for the vacation spots and marker for each city.
locations = hotel_df[["Lat", "Lng"]]

# Create the marker layer
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)

# Set the Size of the Map
figure_layout = {
    'width': '600px',
    'height': '600px',
    'border': '1px solid black'}


fig = gmaps.figure(layout=figure_layout)
city_1_2_city_2_via_two_cities = gmaps.directions_layer(
        city_1_coord, city_4_coord, waypoints=[city_2_coord, city_3_coord], show_markers=False,
        travel_mode='DRIVING')
fig.add_layer(city_1_2_city_2_via_two_cities)
fig.add_layer(marker_layer)
fig
```
